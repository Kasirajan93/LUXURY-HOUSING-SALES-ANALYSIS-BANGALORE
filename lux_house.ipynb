```python
df.drop_duplicates(inplace=True) # Remove all duplicate rows from  df and update it in place
```

```python
df.sample(10) # Randomly sample 10 rows from df for inspection
```

```python
df['Ticket_Price_Cr']=df['Ticket_Price_Cr'].str.replace('Cr','',regex=False).str.strip().str.replace('Rs','').str.replace('₹','')
# Clean the 'Ticket_Price_Cr' column by removing unwanted characters like 'Cr', 'Rs', and '₹', and then strip extra spaces
df['Ticket_Price_Cr']=pd.to_numeric(df['Ticket_Price_Cr'],errors='coerce')
# Convert the cleaned 'Ticket_Price_Cr' column to numeric values; invalid parsing will be set as NaN
```

```python
df['Ticket_Price_Cr'] # Display the cleaned and converted 'Ticket_Price_Cr' column to verify the changes
```

```python
df['Unit_Size_Sqft'] = df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].mean())
# Fill missing values in 'Unit_Size_Sqft' with the mean of the column
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].mean())
# Fill missing values in 'Ticket_Price_Cr' with the mean of the column
df['Amenity_Score'] = df['Amenity_Score'].fillna(df['Amenity_Score'].median())
# Fill missing values in 'Amenity_Score' with the median of the column
df['Buyer_Comments'] = df['Buyer_Comments'].fillna("No Comments")
# Fill missing values in 'Buyer_Comments' with the placeholder text "No Comments"
```

```python
df
```

```python
df.dtypes
```

```python
# normalize text columns(list of columns containing text data )
norm_column = ['Micro_Market', 'Developer_Name', 'Configuration',
             'Transaction_Type', 'Buyer_Type', 'Possession_Status',
             'Sales_Channel', 'NRI_Buyer']
```

```python
# for loop through each column in the list of text columns
for col in norm_column:
    df[col]=df[col].str.strip().str.title()
# remove spaces and convert text to title case    
```

```python
# Calculate price per square foot by converting 'Ticket_Price_Cr' from crores to base currency
# and dividing it by 'Unit_Size_Sqft'
df["Price_per_Sqft"] = df["Ticket_Price_Cr"] * 10000000 / df["Unit_Size_Sqft"]
```

```python
# Convert the 'Purchase_Quarter' column to datetime format for easier date manipulation
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'])
# Extract the quarter (1 to 4) from the 'Purchase_Quarter' datetime column and store it in 'Quarter_Number'
df['Quarter_Number'] = df['Purchase_Quarter'].dt.quarter
# Display the 'Quarter_Number' column to confirm the extracted quarter values
df['Quarter_Number']
```

```python
df['Booking_Flag'] = df['Transaction_Type'].str.lower().map({'primary': 1, 'secondary': 0}) 
# Create a new column 'Booking_Flag' by mapping 'Transaction_Type' values to binary:
# Convert 'Transaction_Type' to lowercase and map 'primary' to 1 and 'secondary' to 0
```

```python
df['Booking_Flag'] # Display the 'Booking_Flag' column to verify that the mapping has been applied correctly
```

```python
df.rename(columns={"Developer_Name": "Builder"}, inplace=True) 
# Rename the column 'Developer_Name' to 'Builder' for better clarity or standardization
# Use inplace=True to apply the change directly to the dataframe without creating a new one
```
